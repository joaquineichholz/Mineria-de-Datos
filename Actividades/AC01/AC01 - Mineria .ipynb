```python
import numpy as np
from numpy import linalg 
import random
import matplotlib.pyplot as pyplot

import mpl_toolkits
from mpl_toolkits.mplot3d import Axes3D
```

```python
mean = [3, 8, 7]
A = np.array([[3,  4,  9], [4,  1, 13], [9, 13,  6]])
cov = np.cov(A)
```

```python
# normalizar los datos para que todos tengan el mismo peso MUY IMPORTANTE
data = np.random.multivariate_normal(mean, cov, 1000).T

mudata = np.mean(data)

sddata = np.std(data)

data = (data - mudata)/sddata

x, y, z = data

mudatanorm = np.mean(data)
```

```python
# Plotear los datos normalizados 
random.shuffle(x)
random.shuffle(y)
random.shuffle(z)

fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z)
pyplot.show()
```

```python
## calcular covarianza de los datos muestrales 
cov2 = np.cov(data)

# calular los valores propios, tambien sacamos los vectoes
v = linalg.eig(cov2)
```

```python
v1 = v[1][1]
v2 = v[1][2]
```

```python
fig = pyplot.figure()
ax = Axes3D(fig)
ax.scatter(x, y, z)

# mudatanorm es de los datos, para que se vea mejor hay que hacerlo por eje 
ax.quiver(mudatanorm, mudatanorm, mudatanorm, *v[1], color='r')

pyplot.show()


```
